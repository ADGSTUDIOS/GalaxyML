```python
df = pd.read_csv("Balanced_Data.csv")
```

```python
df.head()
```

```python
X = df.iloc[:,0:4]
y = pd.get_dummies(df['Target'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10, stratify=y)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler(feature_range=(-1,1)).fit(X_train)
X_train = scaling.transform(X_train)
X_test = scaling.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(algorithm = 'auto', n_neighbors=8, p= 1)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

# matrics to test the model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
from sklearn import metrics
print("Test - Accuracy :", metrics.accuracy_score(y_test, y_pred))
ml.plot_confusion_matrix(y_test,y_pred,'Confusion Matrix of Model','+','-')
print("Test - classification report : ") 
print(metrics.classification_report(y_test, y_pred))
```

```python
dataset = pd.DataFrame({'u-g': X_test[:, 0], 'g-r': X_test[:, 1], 'r-i': X_test[:, 2], 'i-z': X_test[:, 3]})
Target = pd.DataFrame([y_pred])
CLASS = Target.transpose()
CLASS.columns=['Target']
knn_classifier = pd.concat([dataset,CLASS],axis=1)
knn_classifier
```

```python
parameters = knn_classifier.values
c_train_star = []
u_g_train_star = []
g_r_train_star = []
r_i_train_star = []
i_z_train_star = []
u_r_train_star = []
r_z_train_star = []
r_train_star = []
for p in parameters:
    if p[-1] ==1:
        u_g_train_star.append(p[0])
        g_r_train_star.append(p[1])
        r_i_train_star.append(p[2])
        i_z_train_star.append(p[3])
        c_train_star.append(p[4])
    else:
        pass


c_train_galaxy = []
u_g_train_galaxy = []
g_r_train_galaxy = []
r_i_train_galaxy = []
i_z_train_galaxy = []
u_r_train_galaxy = []
r_z_train_galaxy = []
for p in parameters:
    if p[-1] ==0:
        u_g_train_galaxy.append(p[0])
        g_r_train_galaxy.append(p[1])
        r_i_train_galaxy.append(p[2])
        i_z_train_galaxy.append(p[3])
        c_train_galaxy.append(p[4])
    else:
        pass
```
